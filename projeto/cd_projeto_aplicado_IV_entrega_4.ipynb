# **Introdução**

A educação constitui um dos pilares fundamentais para o desenvolvimento social, econômico e cultural de uma nação, desempenhando papel central na promoção da cidadania, na redução das desigualdades e na ampliação das oportunidades individuais e coletivas. No contexto brasileiro, entretanto, o sistema educacional ainda enfrenta desafios significativos relacionados à desigualdade de acesso, permanência e qualidade do ensino. Tais dificuldades se manifestam de forma heterogênea entre as unidades federativas, refletindo as distintas condições socioeconômicas, culturais e estruturais das regiões do país.

Nesse cenário, o estudo dos fluxos escolares, em especial das taxas de promoção, repetência e evasão, torna-se essencial para compreender o funcionamento e a eficiência do sistema educacional. Essas variáveis permitem analisar não apenas o desempenho dos estudantes, mas também os fatores institucionais e contextuais que influenciam a trajetória escolar. As taxas de transição, disponibilizadas pelo Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (INEP)¹, configuram-se como indicadores valiosos nesse processo, pois expressam a proporção de alunos que avançam para a série subsequente, possibilitando inferências sobre a continuidade, a interrupção ou o abandono dos estudos.

Considerando a relevância dessas informações para a formulação de políticas públicas, o presente projeto tem como objetivo analisar a evolução das taxas de transição por estado brasileiro ao longo dos anos, com ênfase no ensino médio. Busca-se, assim, identificar padrões, rupturas e disparidades regionais, contribuindo para o debate sobre a equidade educacional e para a construção de estratégias voltadas à melhoria da permanência e da progressão dos estudantes na educação básica.



# **Referencial Teórico**

Alternativas metodológicas frequentemente empregadas em estudos de evasão envolvem modelos de séries temporais univariadas (como ARIMA), que permitem capturar tendência, autocorrelação e possíveis rupturas estruturais (SAVIANI, 2008). A vantagem desses modelos é explorar a dinâmica interna da própria série e projetar valores futuros. A limitação reside no fato de que eles não comparam diferentes unidades (por exemplo, estados) e não explicam variações que venham de fatores externos.

Outra abordagem é a utilização de modelos em painel com variáveis explicativas, nos quais a taxa de transição ou abandono é explicada por indicadores socioeconômicos, institucionais ou contextuais. A vantagem é incorporar variáveis externas explicativas; a limitação é que muitos modelos de painel estático não capturam bem persistência temporal ou rupturas estruturais, exigindo especificações mais complexas (painel dinâmico) para tratar essas questões(INEP, 2023).

Dentro desse panorama, (BARBOSA, 2023)² realiza estudo estadual de evasão usando o método de controle sintético para avaliar o impacto de políticas públicas de combate à evasão no ensino médio nos estados de Alagoas e Goiás, baseada em dados do Censo/INEP (BARBOSA, 2023)². Essa abordagem combina dimensão temporal e comparativa entre unidades, mas exige séries longas e condições de comparabilidade para gerar resultados confiáveis.

Para este trabalho, optaremos por uma abordagem de séries temporais interestaduais focada em taxas de transição estaduais. Essa escolha possibilita analisar trajetórias ao longo dos anos, identificar rupturas e projetar cenários futuros, sem depender de variáveis externas não disponíveis em escala uniforme.

# **Pipeline da Solução**

1. **Coleta de dados**  
   A coleta de dados será realizada a partir de informações disponibilizadas pelo Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (INEP), especificamente das taxas de transição escolar, indicadores que expressam a proporção de estudantes que avançam de uma série para a seguinte, permanecem na mesma série (repetência) ou abandonam o percurso escolar (evasão).

2. **Limpeza / pré-processamento**  
   Padronizar estados e anos.
            Tratar valores faltantes e outliers.
            A etapa de limpeza e pré-processamento tem como objetivo garantir a consistência, integridade e comparabilidade dos dados antes da realização das análises descritivas e inferenciais. Esse processo é fundamental para reduzir ruídos, eliminar inconsistências e assegurar que os resultados obtidos reflitam adequadamente a realidade dos fluxos escolares analisados.

3. **Análise exploratória**  
   Visualizar séries, estatísticas descritivas e padrões. A etapa de análise exploratória dos dados tem como finalidade compreender o comportamento das taxas de transição escolar ao longo do tempo e entre as diferentes unidades federativas. Trata-se de uma fase essencial para identificar padrões, tendências, rupturas e possíveis inconsistências nos dados antes da aplicação de métodos estatísticos mais avançados.

4. **Modelagem**  
   Ajustar modelos de série temporal para estados ou grupos.
            A etapa de modelagem estatística tem como propósito compreender a dinâmica temporal das taxas de transição escolar e identificar padrões estruturais, rupturas ou tendências persistentes no comportamento dessas séries ao longo dos anos. A partir dos dados previamente limpos e organizados, serão ajustados modelos de séries temporais, aplicados de forma individual para cada unidade federativa ou, quando conveniente, para grupos de estados com características educacionais semelhantes.

5. **Previsão**  
   Projetar últimas taxas para anos futuros.
            A etapa de previsão tem como objetivo projetar o comportamento futuro das taxas de transição escolar a partir dos modelos de séries temporais ajustados para cada estado ou grupo de estados. Essa fase visa identificar tendências esperadas, potenciais rupturas e cenários prospectivos que possam orientar o planejamento de políticas educacionais e a alocação de recursos públicos. Com base nos modelos selecionados na etapa de modelagem (como ARIMA, SARIMA ou suavização exponencial), serão realizadas projeções das taxas de transição — incluindo promoção, repetência e evasão — para um horizonte temporal definido, por exemplo, cinco anos à frente do último dado disponível. As previsões serão calculadas utilizando as estimativas dos parâmetros ajustados, levando em consideração a estrutura de autocorrelação e a variabilidade inerente de cada série estadual.

6. **Visualização e produto**  
   Dashboard mínimo.
            Gráficos interativos.
            Relatório com insights.
            A etapa de visualização e produção de resultados tem como objetivo sintetizar os achados da análise de séries temporais e das previsões de forma clara, acessível e interativa, permitindo que diferentes públicos compreendam as tendências, disparidades e padrões emergentes das taxas de transição escolar.

7. **Apresentação final**  
   Entregar relatório final, slides e repositório com os resultados.
